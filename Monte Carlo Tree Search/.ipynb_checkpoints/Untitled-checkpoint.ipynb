```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from mcts import MonteCarloTreeSearch
from games.tictactoe import TicTacToeState
```

```python
mcts = MonteCarloTreeSearch(c_param=1.0, computational_budge=18)
```

```python
init_state = np.zeros((3, 3))
# init_state[1, 2] = 1
# init_state[1, 0] = -1
# init_state[0, 2] = 1
root_state = TicTacToeState(init_state, 'o')
root_state.show()
```

```python
best_child, histories = mcts.search(root_state)
```

```python
size = 1.8
rols, cols = len(histories), 10

plt.figure(figsize=(cols * size, rols * size))

for i, history in enumerate(histories):
    for j, (state, title) in enumerate(history):
        idx = i * cols + j
        plt.subplot(rols, cols, idx+1)
        
        if isinstance(state, TicTacToeState):
            state.show()
        else:
            state.state.show()
        plt.title(title)
```
